```python
plt.plot(x,y,ls="--",c='y',linewidth=10)
font1 = {'family':'serif','color':'blue','size':21}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Sports Watch Data", fontdict = font1,loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
```

```python
plt.plot(x,y,ls="--",c='y',linewidth=10)
font1 = {'family':'serif','color':'blue','size':21}
font2 = {'family':'serif','color':'darkred','size':15}
plt.grid(color = 'red', linestyle = ':', linewidth = 3)
plt.title("Sports Watch Data", fontdict = font1,loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
```

```python
x=data['likes']
y=data['dislikes']
plt.suptitle('views')
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Like&Dis')
x=data['likes']
y=data['rating']
plt.subplot(2,1,2)
plt.plot(x,y)
plt.title('Like&Rating')
plt.show()
```

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
y1=[0.3,0.4,0.2,0.1]
myexplode = [0.2, 0.1,0, 0]
mycolors = ["red", "hotpink", "b", "#4CAF50"]
plt.pie(y1, labels = mylabels, explode = myexplode,startangle = 90,colors = mycolors, shadow = True)
plt.legend(title = "Four Fruits:")
plt.show() 
```

```python
from matplotlib import pyplot as plt  
from matplotlib import style  
style.use('ggplot')  
x = [16, 8, 10]  
y = [8, 16, 6]  
x2 = [8, 15, 11]  
y2 = [6, 15, 7]  
plt.plot(x, y, 'r', label='line one', linewidth=5)  
plt.plot(x2, y2, 'm', label='line two', linewidth=5)  
plt.title('Epic Info')  
fig = plt.figure()  
plt.ylabel('Y axis')  
plt.xlabel('X axis')  
plt.legend()  
plt.grid(True, color='k')  
plt.show()  
```
